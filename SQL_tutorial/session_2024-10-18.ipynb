```python
from sqlalchemy import (
    create_engine, inspect, text, select,
    MetaData, Table, and_, or_, desc, asc, func,
    )

import pandas as pd
```

```python
# dialect + driver :// username : password @ host : port / dbname

conn_str = "sqlite+pysqlite://///workspaces/python-d2i-beginners/SQL_tutorial/chinook.db"

chinook_engine = create_engine(conn_str)

with chinook_engine.connect() as conn:
    inspection = inspect(chinook_engine)
    print(inspection.get_table_names())
```

```python
conn_str = "sqlite+pysqlite://///workspaces/python-d2i-beginners/SQL_tutorial/gravity.db"

gravity_engine = create_engine(conn_str)

with gravity_engine.connect() as conn:
    inspection = inspect(gravity_engine)
    print(inspection.get_table_names())
```

```python
chinook_metadata = MetaData()
print(chinook_metadata)

tracks = Table('tracks', chinook_metadata, autoload_with=chinook_engine)
#tracks.c.keys()
repr(tracks)
```

```python
# Make an empty metadata object for the gravity db
# Make a table for the book db table
# get the names of the columns in the book table

gravity_metadata = MetaData()
books = Table('book', gravity_metadata, autoload_with=gravity_engine)
books.c.keys()
```

```python
with chinook_engine.connect() as conn:
    # statement
    stmt = "select * from tracks"
    # results proxy
    result_proxy = conn.execute(text(stmt))
    # fetch results
    result = result_proxy.fetchmany(size=20)

all_tracks = pd.DataFrame(result)
all_tracks
```

```python
with gravity_engine.connect() as conn:
    # statement
    stmt = "select * from book order by title desc"
    # results proxy
    result_proxy = conn.execute(text(stmt))
    # fetch results
    result = result_proxy.fetchmany(size=20)

all_books = pd.DataFrame(result)
all_books
```

```python
with chinook_engine.connect() as conn:
    stmt = select(tracks)
    result_proxy = conn.execute(stmt)
    result = result_proxy.fetchmany(10)

tracks_df = pd.DataFrame(result)
tracks_df
```

```python
# Get the first ten rows of the books table into a dataframe

with gravity_engine.connect() as conn:
    stmt = select(books)
    result_proxy = conn.execute(stmt)
    result = result_proxy.fetchmany(10)

books_df = pd.DataFrame(result)
books_df
```

```python
with chinook_engine.connect() as conn:
    # select * from tracks where AlbumId = 1
    stmt = select(tracks).where(tracks.columns.AlbumId == 1)
    result_proxy = conn.execute(stmt)
    result = result_proxy.fetchall()

album_id_1 = pd.DataFrame(result)
album_id_1
```

```python
# Find rows with books from Gravity with titles 'Working Effectively with Legacy Code' and "The Basic Works of Aristotle"
with gravity_engine.connect() as conn:
    stmt = select(books).where(or_(books.columns.title == 'Working Effectively with Legacy Code',
                                   books.columns.title == "The Basic Works of Aristotle"))
    qry = conn.execute(stmt)
    results = qry.fetchall()

books_df = pd.DataFrame(results)
books_df
```

```python
with chinook_engine.connect() as conn:
    stmt = select(tracks).order_by(tracks.columns.Composer.desc(), tracks.columns.Name)
    qry = conn.execute(stmt)
    results = qry.fetchmany(10)

tracksbycomposer_df = pd.DataFrame(results)
tracksbycomposer_df
```

```python
# order books by pagecount to find ten longest
with gravity_engine.connect() as conn:
    stmt = select(books).order_by(books.columns.num_pages.desc())
    qry = conn.execute(stmt)
    results = qry.fetchmany(10)

longestbooks_df = pd.DataFrame(results)
longestbooks_df
```

```python
with chinook_engine.connect() as conn:
    stmt = select(func.sum(tracks.columns.Milliseconds))
    results = conn.execute(stmt).scalar()
results
```

```python

with gravity_engine.connect() as conn:
    stmt = select(func.sum(books.columns.num_pages))
    result = conn.execute(stmt).scalar()
print(result)
```
