````python
child_identifiers = pd.read_csv(filepath_ci)
child_identifiers.info()
child_identifiers
````

````python
datecols = ['PersonBirthDate', 'ExpectedPersonBirthDate', 'PersonDeathDate']

#TODO: List-based implicit iteration???
child_identifiers[datecols] = pd.to_datetime(
        child_identifiers[datecols], 
        format="%Y-%m-%d",
        errors="coerce"
    )

child_identifiers.info()
child_identifiers
````

````python

for colname in ['PersonBirthDate', 'ExpectedPersonBirthDate', 'PersonDeathDate']:
    child_identifiers[colname] = pd.to_datetime(
        child_identifiers[colname], 
        format="%Y-%m-%d",
        errors="coerce"
    )
````

````python
child_identifiers.info()
child_identifiers[child_identifiers['PersonDeathDate'].notna()]
````

````python
#child_identifiers['Age of Death'] = child_identifiers['PersonDeathDate'] - child_identifiers['PersonBirthDate']
child_identifiers['Age of Death'] = child_identifiers['PersonBirthDate'] - child_identifiers['PersonDeathDate']
# child_identifiers[child_identifiers['Age of Death'].notna()]
child_identifiers['Age of Death'] = child_identifiers['Age of Death'] /  pd.Timedelta(365.25, 'd')
child_identifiers['Age of Death'] = child_identifiers['Age of Death'].astype('int', errors='ignore')
child_identifiers[child_identifiers['PersonDeathDate'].notna()]
````

````python
today = pd.to_datetime(str(date.today()), format='%Y-%m-%d')
child_identifiers['Age today'] = today - child_identifiers['PersonBirthDate']
# child_identifiers[child_identifiers['Age of Death'].notna()]
child_identifiers['Age today'] = child_identifiers['Age today'] / pd.Timedelta(365.25, 'd')
child_identifiers.info()
child_identifiers['Age today'] = child_identifiers['Age today'].astype('int', errors='ignore')
child_identifiers[child_identifiers['Age today'].notna()]
````
